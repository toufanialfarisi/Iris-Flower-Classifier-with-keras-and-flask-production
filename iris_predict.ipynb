```python
# import library keras
import keras

# import library pandas
import pandas as pd

# kita gunakan metode load_model untuk load model NN yang sudah kita training
from keras.models import load_model

print('done')
```

```python
# Load model NN yang sudah ditraining
iris_model = load_model('iris_model.h5')

print('done')
```

```python
# load label class dari model NN
import pickle
file = open('label.pickle', 'rb').read()
lb = pickle.loads(file)

# lihat apakah berhasil atau tidak
lb.classes_
```

```python
# import library numpy
import numpy as np

# panggil file data_testing yang sudah dieksport tadi
data_testing = pd.read_csv('data_testing.csv', index_col=0)

# lakukan method prediksi lalu hasil prediksi disimpan ke dala variabel pred
pred = iris_model.predict(data_testing).argmax(axis=1)

# siapkan variabel kosong untuk menampung (stacking) hasil-hasil prediksi tersebut
hasil_pred = []

# memulai stacking
for i in pred:
    label_name = lb.classes_[i]
    hasil_pred.append(label_name)
    
# ubah format array ke dalam array numpy agar memudahkan dalam manipulasi/komputasi
prediksi = np.array(hasil_pred)

# memasukan feature hasil_prediksi ke dalam dataset agar lebih memudahkan
data_testing.insert(loc=4, column='hasil_prediksi', value=prediksi)

# lihat hasil prediksinya
data_testing
```

```python
# coba sekarang buat data input sembarang
test = [7.7,2.6,6.9,2.3]

# tambah dimensi data variabel test menjadi 2-dimensional array
dtt = np.expand_dims(test, axis=1)

# lakukan prediksi seperti biasa dan menampung hasil prediksi ke dalam variabel p
p = iris_model.predict(dtt.T).argmax(1)[0]

# melihat hasil prediksinya
lb.classes_[p]
```
